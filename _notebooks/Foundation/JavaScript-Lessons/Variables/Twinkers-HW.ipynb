```python
%%html

<h2>Homework Output</h2>
<div id="output"></div>

<script>
  document.getElementById("output").innerText = ""; // Clear output
  (() => {

  // Part A - Creating Variables
  // 1. Create a variable called name and store your first name in it
  let name = "Precia";
  console.log(name);
  document.getElementById("output").innerText += "Name: " + name;

  // 2. Create two variables age and city
  const age = 14;
  let city = "San Diego";
  console.log("I am " + age + " years old and I live in " + city + ".");
  document.getElementById("output").innerText += "\nI am " + age + " years old and I live in " + city + ".";

  // 3. Create a variable isStudent
  const isStudent = true;
  console.log("Is student: " + isStudent);
  document.getElementById("output").innerText += "\nIs student: " + isStudent;

  // Part B - Numbers & Strings
  // 4. Create two number variables and print operations
  const num1 = 10;
  const num2 = 5;
  
  const sum = num1 + num2;
  const difference = num1 - num2;
  const product = num1 * num2;
  const quotient = num1 / num2;
  
  console.log("Numbers: " + num1 + " and " + num2);
  console.log("Sum: " + sum);
  console.log("Difference: " + difference);
  console.log("Product: " + product);
  console.log("Quotient: " + quotient);
  
  document.getElementById("output").innerText += "\n\nNumbers: " + num1 + " and " + num2;
  document.getElementById("output").innerText += "\nSum: " + sum;
  document.getElementById("output").innerText += "\nDifference: " + difference;
  document.getElementById("output").innerText += "\nProduct: " + product;
  document.getElementById("output").innerText += "\nQuotient: " + quotient;
  
  // 5. Create a favoriteFood variable
  const favoriteFood = "pizza";
  console.log("My favorite food is " + favoriteFood + ".");
  document.getElementById("output").innerText += "\n\nMy favorite food is " + favoriteFood + ".";
  
  // Part C - Practice Problems
  document.getElementById("output").innerText += "\n\n--- Part C - Practice Problems ---";
  
  // 6. Swap the values of two variables
  let x = 7;
  let y = 3;
  console.log("Before swap: x = " + x + ", y = " + y);
  document.getElementById("output").innerText += "\n\nBefore swap: x = " + x + ", y = " + y;
  
  // Swapping using a temporary variable
  let temp = x;
  x = y;
  y = temp;
  
  console.log("After swap: x = " + x + ", y = " + y);
  document.getElementById("output").innerText += "\nAfter swap: x = " + x + ", y = " + y;
  
  // 7. Create a fullName variable by joining two strings
  const firstName = "FirstName";
  const lastName = "LastName";
  const fullName = firstName + " " + lastName;
  
  console.log("Full name: " + fullName);
  document.getElementById("output").innerText += "\n\nFull name: " + fullName;
  
  // 8. Convert temperature C = 25 into Fahrenheit
  const celsius = 25;
  const fahrenheit = (celsius * 9/5) + 32;
  
  console.log(celsius + "¬∞C = " + fahrenheit + "¬∞F");
  document.getElementById("output").innerText += "\n\n" + celsius + "¬∞C = " + fahrenheit + "¬∞F";
  
  // 9. Create a variable score and print "Pass" or "Fail"
  // For Extra Credit: Use random number from 1-100 instead of hardcoding 85
  const score = Math.floor(Math.random() * 100) + 1; // Random number between 1-100
  const result = score >= 50 ? "Pass" : "Fail";
  
  console.log("Score: " + score + " - " + result);
  document.getElementById("output").innerText += "\n\nScore: " + score + " - " + result;
  
  // 10. Write a program that asks for name and age
  document.getElementById("output").innerText += "\n\n--- User Input Simulation ---";
  // Since we can't use actual prompt in a Jupyter Notebook cell, we'll simulate it:
  const simulatedName = "Precia";
  const simulatedAge = 14;
  
  console.log("Hello " + simulatedName + ", you are " + simulatedAge + " years old.");
  document.getElementById("output").innerText += "\nHello " + simulatedName + ", you are " + simulatedAge + " years old.";
  document.getElementById("output").innerText += "\n(Note: This is simulated as prompt() doesn't work in notebooks)";
  
  // 11. Creative Project
  document.getElementById("output").innerText += "\n\n--- Creative Project: Simple Calculator ---";
  
  // Using 5 variables to create a simple calculator
  const num1Creative = 10;
  const num2Creative = 5;
  const operation = "multiply"; // could be add, subtract, multiply, divide
  let resultCreative;
  let message;
  
  // Calculate based on operation
  if (operation === "add") {
    resultCreative = num1Creative + num2Creative;
    message = num1Creative + " + " + num2Creative + " = " + resultCreative;
  } else if (operation === "subtract") {
    resultCreative = num1Creative - num2Creative;
    message = num1Creative + " - " + num2Creative + " = " + resultCreative;
  } else if (operation === "multiply") {
    resultCreative = num1Creative * num2Creative;
    message = num1Creative + " √ó " + num2Creative + " = " + resultCreative;
  } else if (operation === "divide") {
    resultCreative = num1Creative / num2Creative;
    message = num1Creative + " √∑ " + num2Creative + " = " + resultCreative;
  } else {
    message = "Invalid operation";
  }
  
  console.log(message);
  document.getElementById("output").innerText += "\n" + message;
  document.getElementById("output").innerText += "\n\nCreative project uses 5 variables: num1Creative, num2Creative, operation, resultCreative, and message.";

  })();
</script>
```

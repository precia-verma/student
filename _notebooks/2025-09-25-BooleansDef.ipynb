## üåü What is a Boolean?
A **Boolean** is a data type that can only have two possible values:
- <span style="color:green; font-weight:bold;">True</span>
- <span style="color:red; font-weight:bold;">False</span>

These values are used to answer questions like:
- "Is this condition met?"
- "Has this event occurred?"

---

## üí° Why are Booleans Important?
Booleans allow programs to make decisions! For example, you can use Booleans to check:
- If a user is logged in
- If a number is positive
- If a file exists

---

## üêç Boolean Examples in Python
Let's look at some simple examples and explain each line:

```python
# Example 1: Simple Boolean assignment
is_sunny = True
```
**Explanation:**
- <span style="color:blue;">is_sunny</span> is a variable that stores a Boolean value. Here, it is set to <span style="color:green;">True</span>, meaning it is sunny.

```python
# Example 2: Boolean from a comparison
is_adult = 18 >= 18
```
**Explanation:**
- <span style="color:blue;">18 >= 18</span> checks if 18 is greater than or equal to 18. This is <span style="color:green;">True</span>.
- <span style="color:blue;">is_adult</span> will be assigned the value <span style="color:green;">True</span>.

```python
# Example 3: Using Booleans in an if statement
age = 16
if age >= 18:
    print("You are an adult.")
else:
    print("You are not an adult.")
```
**Explanation:**
- <span style="color:blue;">age</span> is set to 16.
- <span style="color:blue;">if age >= 18:</span> checks if <span style="color:blue;">age</span> is at least 18. This is <span style="color:red;">False</span>.
- Since the condition is <span style="color:red;">False</span>, the code in the <span style="color:purple;">else</span> block runs, printing "You are not an adult."

```python
# Example 4: Boolean operators
is_raining = False
is_cold = True
if is_raining or is_cold:
    print("Bring a jacket!")
```
**Explanation:**
- <span style="color:blue;">is_raining</span> is <span style="color:red;">False</span>, <span style="color:blue;">is_cold</span> is <span style="color:green;">True</span>.
- <span style="color:orange;">or</span> checks if at least one condition is <span style="color:green;">True</span>. Since <span style="color:blue;">is_cold</span> is <span style="color:green;">True</span>, the message "Bring a jacket!" is printed.

---

## üéØ Summary
- Booleans are <span style="color:green;">True</span> or <span style="color:red;">False</span> values.
- They are used for decision making in code.
- You can create Booleans directly, from comparisons, or by combining conditions with operators like <span style="color:orange;">and</span>, <span style="color:orange;">or</span>, and <span style="color:orange;">not</span>.

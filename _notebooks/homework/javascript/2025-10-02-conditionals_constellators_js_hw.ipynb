```python
// Conditionals Homework

console.log("=== Conditionals Homework ===");

// 1) Temperature check
console.log("1) Temperature Check:");
let temperature = 85;
if (temperature > 80) {
    console.log(`Temperature is ${temperature}¬∞F - It's hot outside!`);
} else {
    console.log(`Temperature is ${temperature}¬∞F - It's nice outside!`);
}

// 2) Grade calculator
console.log("2) Grade Calculator:");
let score = 87;
let grade;
if (score >= 90) {
    grade = "A";
} else if (score >= 80) {
    grade = "B";
} else if (score >= 70) {
    grade = "C";
} else {
    grade = "Needs improvement";
}
console.log(`Score of ${score} = Grade: ${grade}`);

// 3) Function for user score input
console.log("3) Grade Function:");
function calculateGrade(userScore) {
    let userGrade;
    if (userScore >= 90) {
        userGrade = "A - Excellent!";
    } else if (userScore >= 80) {
        userGrade = "B - Good job!";
    } else if (userScore >= 70) {
        userGrade = "C - Keep it up!";
    } else {
        userGrade = "Keep practicing!";
    }
    return userGrade;
}
let testScore = 92;
console.log(`Your grade for ${testScore}: ${calculateGrade(testScore)}`);

// 4) Even or odd number
console.log("4) Even/Odd Check:");
let number = 7;
if (number % 2 === 0) {
    console.log(`${number} is an even number`);
} else {
    console.log(`${number} is an odd number`);
}

// 5) Random number game
console.log("5) Random Number Game:");
let randomNum = Math.floor(Math.random() * 10) + 1;
let isEven = (randomNum % 2 === 0);
console.log(`Random number: ${randomNum}`);
console.log(`It is ${isEven ? 'even' : 'odd'}`);

// 6) Day of the week with switch
console.log("6) Day of the Week:");
let day = "Friday";
let dayMessage;
switch (day) {
    case "Monday":
        dayMessage = "Fresh start!";
        break;
    case "Tuesday":
        dayMessage = "Getting momentum!";
        break;
    case "Wednesday":
        dayMessage = "Hump day - you got this!";
        break;
    case "Thursday":
        dayMessage = "Almost Friday!";
        break;
    case "Friday":
        dayMessage = "TGIF - Weekend's here!";
        break;
    case "Saturday":
        dayMessage = "Saturday vibes!";
        break;
    case "Sunday":
        dayMessage = "Sunday funday!";
        break;
    default:
        dayMessage = "Invalid day!";
}
console.log(`${day}: ${dayMessage}`);

// 7) Weekend check
console.log("7) Weekend Check:");
let isWeekend = (day === "Saturday" || day === "Sunday");
console.log(`${day} is a ${isWeekend ? 'Weekend!' : 'Weekday!'}`);

// 8) Login status
console.log("8) Login Status:");
let loggedIn = true;
if (loggedIn) {
    console.log("Welcome back, friend!");
} else {
    console.log("Please log in to continue");
}

// 9) Weather report
console.log("9) Weather Report:");
let weather = "sunny";
if (weather.toLowerCase() === "raining") {
    console.log("It's raining - grab an umbrella!");
} else {
    console.log("Nice weather - perfect day to go outside!");
}

console.log("Homework Complete!");
```

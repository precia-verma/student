```python
%%html

<div id="outputHW">Results will appear here:<br></div>

<script>

(() => {
    const output = document.getElementById("outputHW");
    
    // Clear the default text
    output.innerHTML = "Homework Results:<br>";
    
    // 1) Temperature Check
    function checkTemperature() {
        output.innerHTML += "<br><strong>1) Temperature Check:</strong><br>";
        let temperature = prompt("Enter a temperature in degrees:");
        temperature = parseFloat(temperature);
        
        if (isNaN(temperature)) {
            output.innerHTML += "Please enter a valid number.<br>";
            return;
        }
        
        if (temperature > 80) {
            output.innerHTML += `Temperature: ${temperature}¬∞ - It's hot outside!<br>`;
        } else {
            output.innerHTML += `Temperature: ${temperature}¬∞ - It's nice outside.<br>`;
        }
    }
    
    // 2 & 3) Score Grade with Input
    function checkScore() {
        output.innerHTML += "<br><strong>2 & 3) Score Grade Check:</strong><br>";
        let score = prompt("Enter your score (0-100):");
        score = parseFloat(score);
        
        if (isNaN(score) || score < 0 || score > 100) {
            output.innerHTML += "Please enter a valid score between 0 and 100.<br>";
            return;
        }
        
        if (score >= 90) {
            output.innerHTML += `Score: ${score} - Grade: A<br>`;
        } else if (score >= 80) {
            output.innerHTML += `Score: ${score} - Grade: B<br>`;
        } else if (score >= 70) {
            output.innerHTML += `Score: ${score} - Grade: C<br>`;
        } else {
            output.innerHTML += `Score: ${score} - Grade: Needs improvement<br>`;
        }
    }
    
    // 4 & 5) Number Check with Random Number
    function checkNumber() {
        output.innerHTML += "<br><strong>4 & 5) Even/Odd Number Check:</strong><br>";
        // Random number between 1-10
        let number = Math.floor(Math.random() * 10) + 1;
        
        if (number % 2 === 0) {
            output.innerHTML += `Number: ${number} - Even Number<br>`;
        } else {
            output.innerHTML += `Number: ${number} - Odd Number<br>`;
        }
    }
    
    // 6 & 7) Day of Week with Switch and Weekend Check
    function checkDay() {
        output.innerHTML += "<br><strong>6 & 7) Day of Week Check:</strong><br>";
        const days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'];
        let dayIndex = prompt(`Enter a day number (0-6):\n0 = Sunday\n1 = Monday\n2 = Tuesday\n3 = Wednesday\n4 = Thursday\n5 = Friday\n6 = Saturday`);
        
        dayIndex = parseInt(dayIndex);
        
        if (isNaN(dayIndex) || dayIndex < 0 || dayIndex > 6) {
            output.innerHTML += "Please enter a valid day number between 0 and 6.<br>";
            return;
        }
        
        let day = days[dayIndex];
        
        // Switch statement for day message
        switch(day) {
            case 'Monday':
                output.innerHTML += "Monday: Start of the work week!<br>";
                break;
            case 'Tuesday':
                output.innerHTML += "Tuesday: Keep the momentum going!<br>";
                break;
            case 'Wednesday':
                output.innerHTML += "Wednesday: Halfway there!<br>";
                break;
            case 'Thursday':
                output.innerHTML += "Thursday: Almost to the weekend!<br>";
                break;
            case 'Friday':
                output.innerHTML += "Friday: TGIF!<br>";
                break;
            case 'Saturday':
                output.innerHTML += "Saturday: Time to relax!<br>";
                break;
            case 'Sunday':
                output.innerHTML += "Sunday: End of the weekend!<br>";
                break;
            default:
                output.innerHTML += "Invalid day.<br>";
        }
        
        // Weekend or weekday check
        if (day === 'Saturday' || day === 'Sunday') {
            output.innerHTML += `${day} is a Weekend!<br>`;
        } else {
            output.innerHTML += `${day} is a Weekday!<br>`;
        }
    }
    
    // 8) Login Status Check
    function checkLogin() {
        output.innerHTML += "<br><strong>8) Login Status Check:</strong><br>";
        let loginStatus = confirm("Are you logged in? (OK = Yes, Cancel = No)");
        
        let loggedIn = loginStatus; // Boolean based on user's selection
        
        if (loggedIn) {
            output.innerHTML += "Welcome back!<br>";
        } else {
            output.innerHTML += "Please log in.<br>";
        }
    }
    
    // 9) Weather Check
    function checkWeather() {
        output.innerHTML += "<br><strong>9) Weather Check:</strong><br>";
        let isRaining = confirm("Is it raining? (OK = Yes, Cancel = No)");
        
        let weather = isRaining ? "raining" : "clear";
        
        if (weather === "raining") {
            output.innerHTML += "It's raining. Don't forget your umbrella!<br>";
        } else {
            output.innerHTML += "Go outside, it's a nice day today!<br>";
        }
    }
    
    // Create buttons for each homework example
    output.innerHTML += `<br>
        <button onclick="checkTemperature()" style="margin: 5px; padding: 5px;">1) Check Temperature</button>
        <button onclick="checkScore()" style="margin: 5px; padding: 5px;">2 & 3) Check Score</button>
        <button onclick="checkNumber()" style="margin: 5px; padding: 5px;">4 & 5) Check Random Number</button>
        <button onclick="checkDay()" style="margin: 5px; padding: 5px;">6 & 7) Check Day of Week</button>
        <button onclick="checkLogin()" style="margin: 5px; padding: 5px;">8) Check Login Status</button>
        <button onclick="checkWeather()" style="margin: 5px; padding: 5px;">9) Check Weather</button>
    `;
    
    // Make the functions available globally
    window.checkTemperature = checkTemperature;
    window.checkScore = checkScore;
    window.checkNumber = checkNumber;
    window.checkDay = checkDay;
    window.checkLogin = checkLogin;
    window.checkWeather = checkWeather;

})();

</script>
```

```python
// Conditionals Homework - Interactive Version

console.log("=== Welcome to Conditionals Homework ===");
console.log("This homework will ask you questions and test your conditional logic!");

// 1) Temperature check
console.log("\n1) Temperature Check:");
let temperature = prompt("What's the temperature outside today? (enter a number)");
temperature = parseFloat(temperature);
if (temperature > 80) {
    console.log(`Temperature is ${temperature}¬∞F - It's hot outside!`);
} else {
    console.log(`Temperature is ${temperature}¬∞F - It's nice outside!`);
}

// 2) Grade calculator  
console.log("\n2) Grade Calculator:");
let score = prompt("What score did you get on your test? (0-100)");
score = parseFloat(score);
let grade;
if (score >= 90) {
    grade = "A";
} else if (score >= 80) {
    grade = "B";
} else if (score >= 70) {
    grade = "C";
} else {
    grade = "Needs improvement";
}
console.log(`Your score of ${score} = Grade: ${grade}`);

// 3) Grade function demonstration
console.log("\n3) Grade Function Demo:");
function calculateGrade(userScore) {
    let userGrade;
    if (userScore >= 90) {
        userGrade = "A - Excellent!";
    } else if (userScore >= 80) {
        userGrade = "B - Good job!";
    } else if (userScore >= 70) {
        userGrade = "C - Keep it up!";
    } else {
        userGrade = "Keep practicing!";
    }
    return userGrade;
}
let anotherScore = prompt("Enter another test score to see the grade function in action:");
anotherScore = parseFloat(anotherScore);
console.log(`Function result for ${anotherScore}: ${calculateGrade(anotherScore)}`);

// 4) Even or odd number
console.log("\n4) Even/Odd Check:");
let number = prompt("Enter any number to check if it's even or odd:");
number = parseInt(number);
if (number % 2 === 0) {
    console.log(`${number} is an even number`);
} else {
    console.log(`${number} is an odd number`);
}

// 5) Random number comparison
console.log("\n5) Random Number Game:");
let randomNum = Math.floor(Math.random() * 10) + 1;
let guess = prompt(`I'm thinking of a number from 1-10. Can you guess it? The number is ${randomNum}. Enter your guess:`);
guess = parseInt(guess);
if (guess === randomNum) {
    console.log(`üéâ Amazing! You guessed ${randomNum} correctly!`);
} else {
    console.log(`Nice try! You guessed ${guess}, but I was thinking of ${randomNum}`);
}
let isEven = (randomNum % 2 === 0);
console.log(`The number ${randomNum} is ${isEven ? 'even' : 'odd'}`);

// 6) Day of the week with switch
console.log("\n6) Day of the Week:");
let day = prompt("What day of the week is it today? (e.g., Monday, Tuesday, etc.)");
let dayMessage;
switch (day) {
    case "Monday":
        dayMessage = "Fresh start!";
        break;
    case "Tuesday":
        dayMessage = "Getting momentum!";
        break;
    case "Wednesday":
        dayMessage = "Hump day - you got this!";
        break;
    case "Thursday":
        dayMessage = "Almost Friday!";
        break;
    case "Friday":
        dayMessage = "TGIF - Weekend's here!";
        break;
    case "Saturday":
        dayMessage = "Saturday vibes!";
        break;
    case "Sunday":
        dayMessage = "Sunday funday!";
        break;
    default:
        dayMessage = "Hmm, that doesn't seem like a valid day!";
}
console.log(`${day}: ${dayMessage}`);

// 7) Weekend check
console.log("\n7) Weekend Check:");
let isWeekend = (day === "Saturday" || day === "Sunday");
console.log(`${day} is a ${isWeekend ? 'Weekend! üéâ' : 'Weekday! üíº'}`);

// 8) Login status
console.log("\n8) Login Status:");
let loginResponse = prompt("Are you logged in? (yes/no)");
let loggedIn = (loginResponse.toLowerCase() === "yes" || loginResponse.toLowerCase() === "y");
if (loggedIn) {
    console.log("Welcome back, friend! üëã");
} else {
    console.log("Please log in to continue üîê");
}

// 9) Weather report
console.log("\n9) Weather Report:");
let weather = prompt("How's the weather today? (sunny, rainy, cloudy, etc.)");
if (weather.toLowerCase().includes("rain")) {
    console.log("It's raining - grab an umbrella! ‚òî");
} else if (weather.toLowerCase().includes("sunny")) {
    console.log("Sunny day - perfect for going outside! ‚òÄÔ∏è");
} else {
    console.log(`${weather} weather - sounds nice! üå§Ô∏è`);
}

console.log("\nüéì Homework Complete! Great job working through all the conditionals!");
```

```python
%%html

<div id="outputHW">Results will appear here:<br></div>

<script>

(() => {

    const output = document.getElementById("outputHW");

    output.innerText += "Welcome to my conditionals homework!\n\n";

    // 1) Temperature check
    let temp = 85;
    if (temp > 80) {
        output.innerText += "1) Temperature is " + temp + "°F - It's hot outside!\n";
    } else {
        output.innerText += "1) Temperature is " + temp + "°F - It's nice outside!\n";
    }

    // 2) Grade calculator
    let myScore = 88;
    let grade;
    if (myScore >= 90) {
        grade = "A";
    } else if (myScore >= 80) {
        grade = "B";
    } else if (myScore >= 70) {
        grade = "C";
    } else {
        grade = "Needs improvement";
    }
    output.innerText += "2) Score of " + myScore + " = Grade: " + grade + "\n";

    // 3) User's grade
    let userScore = prompt("What's your score? (Enter a number 0-100)");
    if (userScore) {
        userScore = Number(userScore);
        let userGrade;
        if (userScore >= 90) {
            userGrade = "A - Excellent!";
        } else if (userScore >= 80) {
            userGrade = "B - Good job!";
        } else if (userScore >= 70) {
            userGrade = "C - Keep it up!";
        } else {
            userGrade = "Keep practicing!";
        }
        output.innerText += "3) Your grade: " + userGrade + "\n";
    }

    // 4) Even or odd number
    let num = 8;
    if (num % 2 === 0) {
        output.innerText += "4) " + num + " is an even number\n";
    } else {
        output.innerText += "4) " + num + " is an odd number\n";
    }

    // 5) Random number game
    let randomNum = Math.floor(Math.random() * 10) + 1;
    let evenOrOdd = (randomNum % 2 === 0) ? "even" : "odd";
    output.innerText += "5) Random number " + randomNum + " is " + evenOrOdd + "!\n";

    // 6) Day of the week
    let today = "Wednesday";
    let dayMessage;
    switch (today) {
        case "Monday":
            dayMessage = "Fresh start!";
            break;
        case "Tuesday":
            dayMessage = "Getting momentum!";
            break;
        case "Wednesday":
            dayMessage = "You got this!";
            break;
        case "Thursday":
            dayMessage = "Almost Friday!";
            break;
        case "Friday":
            dayMessage = "Weekend's here!";
            break;
        case "Saturday":
            dayMessage = "Chill day!";
            break;
        case "Sunday":
            dayMessage = "Rest day!";
            break;
    }
    output.innerText += "6) " + today + ": " + dayMessage + "\n";

    // 7) Weekend check
    let weekendCheck = (today === "Saturday" || today === "Sunday") ? "Weekend!" : "Weekday!";
    output.innerText += "7) " + weekendCheck + "\n";

    // 8) Login status
    let isLoggedIn = true;
    if (isLoggedIn) {
        output.innerText += "8) Welcome back, friend!\n";
    } else {
        output.innerText += "8) Please log in to continue\n";
    }

    // 9) Weather report
    let weather = "sunny";
    if (weather === "raining") {
        output.innerText += "9) It's raining - grab an umbrella!\n";
    } else {
        output.innerText += "9) Nice weather - perfect day to go outside!\n";
    }

    output.innerText += "\nYay!";

})();

</script>
```

```python
%%html

<div id="outputPH1">Results will appear here:<br></div>

<script>

(() => {

    const output = document.getElementById("outputPH1");
    output.innerHTML = "Choose an example to run:<br>";
    
    // Example 1: Mood Checker
    function runMoodExample() {
        output.innerHTML = "MOOD EXAMPLE RESULTS:<br>";
        let mood = prompt("How are you feeling today? (happy, sad, tired, excited, anxious)");
        
        if (mood && mood.toLowerCase() === "happy") {
            console.log("You're happy! Here's some ice cream to celebrate! üç¶");
            output.innerHTML += "You're happy! Here's some ice cream to celebrate! üç¶<br>";
        } else if (mood && mood.toLowerCase() === "sad") {
            console.log("You're feeling sad. Here's some chocolate to cheer you up! üç´");
            output.innerHTML += "You're feeling sad. Here's some chocolate to cheer you up! üç´<br>";
        } else if (mood && mood.toLowerCase() === "tired") {
            console.log("You're tired. How about a coffee to boost your energy? ‚òï");
            output.innerHTML += "You're tired. How about a coffee to boost your energy? ‚òï<br>";
        } else if (mood && mood.toLowerCase() === "excited") {
            console.log("You're excited! Let's celebrate with a cake! üéÇ");
            output.innerHTML += "You're excited! Let's celebrate with a cake! üéÇ<br>";
        } else if (mood && mood.toLowerCase() === "anxious") {
            console.log("You're feeling anxious. Take a deep breath and have some tea. üçµ");
            output.innerHTML += "You're feeling anxious. Take a deep breath and have some tea. üçµ<br>";
        } else {
            console.log("I don't recognize that mood, but have a cookie anyway! üç™");
            output.innerHTML += "I don't recognize that mood, but have a cookie anyway! üç™<br>";
        }
    }
    
    // Example 2: Weather Checker
    function runWeatherExample() {
        output.innerHTML = "WEATHER EXAMPLE RESULTS:<br>";
        let weather = prompt("What's the weather like today? (sunny, rainy, cloudy, snowy, windy)");
        
        if (weather && weather.toLowerCase() === "sunny") {
            console.log("It's sunny outside! Perfect day for a walk!");
            output.innerHTML += "It's sunny outside! Perfect day for a walk!<br>";
        } else if (weather && weather.toLowerCase() === "rainy") {
            console.log("It's rainy today. Don't forget your umbrella!");
            output.innerHTML += "It's rainy today. Don't forget your umbrella!<br>";
        } else if (weather && weather.toLowerCase() === "cloudy") {
            console.log("It's cloudy today. The temperature is mild.");
            output.innerHTML += "It's cloudy today. The temperature is mild.<br>";
        } else if (weather && weather.toLowerCase() === "snowy") {
            console.log("It's snowing! Time to build a snowman!");
            output.innerHTML += "It's snowing! Time to build a snowman!<br>";
        } else if (weather && weather.toLowerCase() === "windy") {
            console.log("It's windy today. Hold onto your hat!");
            output.innerHTML += "It's windy today. Hold onto your hat!<br>";
        } else {
            console.log("I don't recognize that weather. Whatever it is, stay safe!");
            output.innerHTML += "I don't recognize that weather. Whatever it is, stay safe!<br>";
        }
    }
    
    // Example 3: Age Checker
    function runAgeExample() {
        output.innerHTML = "AGE EXAMPLE RESULTS:<br>";
        
        // Create a function to test the age conditional multiple times
        function checkAge(age) {
            if (age >= 65) {
                return `Age ${age}: You are a senior citizen. Enjoy your retirement!`;
            } else if (age >= 18) {
                return `Age ${age}: You are an adult. Time to pay bills!`;
            } else if (age >= 13) {
                return `Age ${age}: You are a teenager. Enjoy your youth!`;
            } else if (age >= 5) {
                return `Age ${age}: You are a child. Go play outside!`;
            } else {
                return `Age ${age}: You are a toddler. Time for a nap!`;
            }
        }
        
        // Option to use random ages or user input
        let useRandom = confirm("Do you want to generate random ages? Click OK for random ages, or Cancel to enter your own age.");
        
        if (useRandom) {
            // Generate and check 5 random ages
            for (let i = 0; i < 5; i++) {
                let randomAge = Math.floor(Math.random() * 100); // Random age between 0-99
                console.log(checkAge(randomAge));
                output.innerHTML += checkAge(randomAge) + "<br>";
            }
        } else {
            // Use user input
            let userAge = prompt("Enter your age:");
            if (userAge !== null && !isNaN(parseInt(userAge))) {
                console.log(checkAge(parseInt(userAge)));
                output.innerHTML += checkAge(parseInt(userAge)) + "<br>";
            } else {
                console.log("Invalid age entered.");
                output.innerHTML += "Invalid age entered.<br>";
            }
        }
    }
    
    // Create buttons for the different examples
    output.innerHTML += `
        <button onclick="runMoodExample()" style="margin: 5px; padding: 5px;">Run Mood Example</button>
        <button onclick="runWeatherExample()" style="margin: 5px; padding: 5px;">Run Weather Example</button>
        <button onclick="runAgeExample()" style="margin: 5px; padding: 5px;">Run Age Example</button>
    `;
    
    // Make the functions available globally
    window.runMoodExample = runMoodExample;
    window.runWeatherExample = runWeatherExample;
    window.runAgeExample = runAgeExample;

})();

</script>
```

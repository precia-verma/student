```python
// 1) Temperature check
let temperature = 85;
if (temperature > 80) {
  console.log("It's hot outside!");
} else {
  console.log("It's nice outside.");
}

// 2) Grade calculator
let score = 87;
if (score >= 90) {
  console.log("A");
} else if (score >= 80) {
  console.log("B");
} else if (score >= 70) {
  console.log("C");
} else {
  console.log("Needs improvement");
}

// 3) Grade function
function getGrade(userScore) {
  if (userScore >= 90) return "A";
  if (userScore >= 80) return "B";
  if (userScore >= 70) return "C";
  return "Needs improvement";
}
console.log(getGrade(92));

// 4) Even or odd number
let number = 7;
if (number % 2 === 0) {
  console.log("Even Number");
} else {
  console.log("Odd Number");
}

// 5) Random number from 1-10
let randNum = Math.floor(Math.random() * 10) + 1;
if (randNum % 2 === 0) {
  console.log(randNum + " is Even");
} else {
  console.log(randNum + " is Odd");
}

// 6) Day of the week
let day = "Friday";
switch (day) {
  case "Monday":
    console.log("Start of the week!");
    break;
  case "Friday":
    console.log("It's Friday!");
    break;
  default:
    console.log("It's " + day);
}

// 7) Weekend check
if (day === "Saturday" || day === "Sunday") {
  console.log("Weekend!");
} else {
  console.log("Weekday!");
}

// 8) Login status
let loggedIn = true;
if (loggedIn) {
  console.log("Welcome back!");
} else {
  console.log("Please log in.");
}

// 9) Weather report
let weather = "sunny";
if (weather === "raining") {
  console.log("It's raining.");
} else {
  console.log("Go outside, it's a nice day!");
}
```

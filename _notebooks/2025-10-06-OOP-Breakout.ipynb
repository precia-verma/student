## üéÆ Executable OOP Breakout Game

The game below implements the requested feature: **bricks need to be hit twice before breaking completely**. 

### Key Features Implemented:
- **Two-Hit Brick System**: Each brick has a `hits` counter and `maxHits = 2`
- **Visual Damage Feedback**: Bricks darken and show cracks when hit once
- **Progressive Scoring**: Half points for first hit, full points for destruction
- **Complete OOP Architecture**: GameObject base class with Paddle, Ball, Brick, and Game classes

### Controls:
- **Arrow Keys** or **Mouse** to move paddle
- **Spacebar** or **Click** to launch ball
- **R** to restart, **P** to pause
- **+ (Plus)** to increase ball speed
- **- (Minus)** to decrease ball speed

Click the cell below to run the game!

```python
%%html
<div style="text-align: center; background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); padding: 20px; border-radius: 15px; margin: 20px auto; max-width: 920px;">
    <h2 style="color: white; margin-bottom: 20px;">üéÆ Fixed Enhanced Breakout</h2>
    <div style="background: white; padding: 20px; border-radius: 10px; display: inline-block;">
        <canvas id="fixedCanvas" width="850" height="650" style="border: 3px solid #333; border-radius: 10px; background: linear-gradient(180deg, #001a2e 0%, #003d5c 100%); display: block; margin: 0 auto;"></canvas>
        
        <div style="margin-top: 15px; display: flex; justify-content: space-between; font-size: 18px; font-weight: bold;">
            <div style="background: #4CAF50; color: white; padding: 8px 15px; border-radius: 8px;">Score: <span id="fixedScore">0</span></div>
            <div style="background: #f44336; color: white; padding: 8px 15px; border-radius: 8px;">Lives: <span id="fixedLives">3</span></div>
            <div style="background: #2196F3; color: white; padding: 8px 15px; border-radius: 8px;">Level: <span id="fixedLevel">1</span></div>
            <div style="background: #FF9800; color: white; padding: 8px 15px; border-radius: 8px;">Speed: <span id="fixedSpeed">5</span></div>
        </div>
        
        <div style="margin-top: 10px; color: #666; font-size: 14px;">
            Mouse = Move | Click/Space = Launch | +/- = Speed | R = Restart | P = Pause
        </div>
        
        <div style="margin-top: 10px;">
            <button onclick="fixedGame && fixedGame.restart()" style="background: #4c51bf; color: white; border: none; padding: 10px 20px; border-radius: 5px; cursor: pointer; margin: 5px;">Restart</button>
            <button onclick="fixedGame && fixedGame.togglePause()" style="background: #7c3aed; color: white; border: none; padding: 10px 20px; border-radius: 5px; cursor: pointer; margin: 5px;">Pause</button>
        </div>
        
        <div id="fixedStatus" style="margin-top: 10px; color: #333; font-weight: bold;">Loading...</div>
    </div>
</div>

<script>
console.log("Starting fixed enhanced game...");

// Clear existing
if (window.fixedGame) window.fixedGame = null;

function updateFixedStatus(msg) {
    const el = document.getElementById('fixedStatus');
    if (el) el.textContent = msg;
    console.log('Fixed Game:', msg);
}

// Particle for effects
class FixedParticle {
    constructor(x, y, vx, vy, color, life = 60) {
        this.x = x; this.y = y; this.vx = vx; this.vy = vy;
        this.color = color; this.life = life; this.maxLife = life;
    }
    
    update() {
        this.x += this.vx; this.y += this.vy;
        this.vy += 0.1;
        this.life--;
        return this.life > 0;
    }
    
    draw(ctx) {
        const alpha = this.life / this.maxLife;
        ctx.save();
        ctx.globalAlpha = alpha;
        ctx.fillStyle = this.color;
        ctx.fillRect(this.x, this.y, 3, 3);
        ctx.restore();
    }
}

// Enhanced GameObject
class FixedObject {
    constructor(x, y, w, h, color) {
        this.x = x; this.y = y; this.w = w; this.h = h; this.color = color;
    }
    
    draw(ctx) {
        ctx.fillStyle = this.color;
        ctx.fillRect(this.x, this.y, this.w, this.h);
        ctx.strokeStyle = '#333';
        ctx.lineWidth = 1;
        ctx.strokeRect(this.x, this.y, this.w, this.h);
    }
    
    collides(other) {
        return this.x < other.x + other.w && this.x + this.w > other.x &&
               this.y < other.y + other.h && this.y + this.h > other.y;
    }
}

// Fixed Paddle
class FixedPaddle extends FixedObject {
    constructor() {
        super(375, 600, 100, 15, '#00ff88');
        this.targetX = 375;
    }
    
    update() {
        const diff = this.targetX - this.x;
        this.x += diff * 0.15;
        if (this.x < 0) this.x = 0;
        if (this.x + this.w > 850) this.x = 850 - this.w;
    }
    
    moveTo(x) {
        this.targetX = x - this.w / 2;
    }
    
    draw(ctx) {
        ctx.save();
        ctx.shadowColor = this.color;
        ctx.shadowBlur = 10;
        super.draw(ctx);
        ctx.restore();
    }
}

// Fixed Ball with speed control
class FixedBall extends FixedObject {
    constructor() {
        super(425, 580, 12, 12, '#ffff00');
        this.dx = 0; this.dy = 0; this.speed = 5;
        this.launched = false; this.trail = [];
    }
    
    launch() {
        if (!this.launched) {
            this.dx = (Math.random() - 0.5) * 4;
            this.dy = -this.speed;
            this.launched = true;
            updateFixedStatus('Ball launched!');
        }
    }
    
    update(canvas, paddle, particles) {
        if (!this.launched) {
            this.x = paddle.x + paddle.w/2 - this.w/2;
            this.y = paddle.y - this.h - 5;
            return 'waiting';
        }
        
        // Trail effect
        this.trail.push({x: this.x, y: this.y, life: 20});
        this.trail = this.trail.filter(t => t.life-- > 0);
        
        this.x += this.dx; this.y += this.dy;
        
        // Wall bouncing
        if (this.x <= 0 || this.x >= canvas.width - this.w) {
            this.dx = -this.dx;
            this.x = Math.max(0, Math.min(this.x, canvas.width - this.w));
            this.createParticles(particles);
        }
        if (this.y <= 0) {
            this.dy = -this.dy;
            this.y = 0;
            this.createParticles(particles);
        }
        
        if (this.y > canvas.height) return 'lost';
        
        // Paddle collision
        if (this.collides(paddle) && this.dy > 0) {
            const hitPos = (this.x + this.w/2 - paddle.x) / paddle.w;
            const angle = (hitPos - 0.5) * Math.PI * 0.6;
            this.dx = Math.sin(angle) * this.speed;
            this.dy = -Math.cos(angle) * this.speed;
            this.y = paddle.y - this.h;
            this.createParticles(particles);
        }
        
        return 'active';
    }
    
    createParticles(particles) {
        for (let i = 0; i < 5; i++) {
            const vx = (Math.random() - 0.5) * 4;
            const vy = (Math.random() - 0.5) * 4;
            particles.push(new FixedParticle(this.x + this.w/2, this.y + this.h/2, vx, vy, this.color));
        }
    }
    
    // *** HOMEWORK: SPEED CONTROL METHOD ***
    changeSpeed(delta) {
        this.speed += delta;
        this.speed = Math.max(2, Math.min(10, this.speed));
        
        if (this.launched) {
            const angle = Math.atan2(this.dx, -this.dy);
            this.dx = Math.sin(angle) * this.speed;
            this.dy = -Math.cos(angle) * this.speed;
        }
        
        document.getElementById('fixedSpeed').textContent = this.speed;
        updateFixedStatus(`Speed: ${this.speed}`);
        console.log(`*** Ball speed changed to: ${this.speed} ***`);
    }
    
    draw(ctx) {
        // Draw trail
        this.trail.forEach((t, i) => {
            const alpha = t.life / 20;
            ctx.save();
            ctx.globalAlpha = alpha * 0.5;
            ctx.fillStyle = this.color;
            ctx.fillRect(t.x, t.y, this.w * alpha, this.h * alpha);
            ctx.restore();
        });
        
        // Draw ball with glow
        ctx.save();
        ctx.shadowColor = this.color;
        ctx.shadowBlur = 15;
        super.draw(ctx);
        ctx.restore();
    }
    
    reset(paddle) {
        this.launched = false;
        this.dx = 0; this.dy = 0; this.trail = [];
        this.x = paddle.x + paddle.w/2 - this.w/2;
        this.y = paddle.y - this.h - 5;
    }
}

// Fixed Brick with TWO-HIT system
class FixedBrick extends FixedObject {
    constructor(x, y, type) {
        const colors = ['#ff4444', '#44ff44', '#4444ff', '#ffff44', '#ff44ff'];
        super(x, y, 80, 25, colors[type - 1]);
        this.type = type;
        this.hits = 0;
        this.maxHits = 2; // *** HOMEWORK: TWO-HIT FEATURE ***
        this.points = type * 10;
        this.originalColor = this.color;
        this.destroyed = false;
    }
    
    hit(particles) {
        this.hits++;
        console.log(`*** Brick hit! ${this.hits}/${this.maxHits} hits ***`);
        
        // Create particles
        for (let i = 0; i < 8; i++) {
            const vx = (Math.random() - 0.5) * 6;
            const vy = (Math.random() - 0.5) * 6;
            particles.push(new FixedParticle(this.x + this.w/2, this.y + this.h/2, vx, vy, this.color));
        }
        
        if (this.hits === 1) {
            // First hit - darken
            this.color = this.darkenColor(this.originalColor);
            return Math.floor(this.points / 2);
        }
        
        if (this.hits >= this.maxHits) {
            // Second hit - destroy
            this.destroyed = true;
            return this.points;
        }
        
        return 0;
    }
    
    darkenColor(color) {
        const r = Math.floor(parseInt(color.slice(1, 3), 16) * 0.4);
        const g = Math.floor(parseInt(color.slice(3, 5), 16) * 0.4);
        const b = Math.floor(parseInt(color.slice(5, 7), 16) * 0.4);
        return `#${r.toString(16).padStart(2, '0')}${g.toString(16).padStart(2, '0')}${b.toString(16).padStart(2, '0')}`;
    }
    
    draw(ctx) {
        super.draw(ctx);
        
        // Damage cracks
        if (this.hits === 1) {
            ctx.strokeStyle = '#000';
            ctx.lineWidth = 2;
            ctx.beginPath();
            ctx.moveTo(this.x + 15, this.y + 5);
            ctx.lineTo(this.x + 35, this.y + 15);
            ctx.moveTo(this.x + 50, this.y + 8);
            ctx.lineTo(this.x + 65, this.y + 18);
            ctx.stroke();
        }
        
        // Type number
        ctx.fillStyle = '#fff';
        ctx.font = '12px Arial';
        ctx.textAlign = 'center';
        ctx.fillText(this.type.toString(), this.x + this.w/2, this.y + this.h/2 + 4);
    }
}

// Fixed Game class
class FixedGame {
    constructor() {
        this.canvas = document.getElementById('fixedCanvas');
        if (!this.canvas) {
            updateFixedStatus('ERROR: Canvas not found!');
            return;
        }
        
        this.ctx = this.canvas.getContext('2d');
        if (!this.ctx) {
            updateFixedStatus('ERROR: Context failed!');
            return;
        }
        
        this.score = 0; this.lives = 3; this.level = 1;
        this.paused = false; this.gameOver = false;
        this.particles = [];
        
        this.init();
        this.setupControls();
        this.start();
        
        updateFixedStatus('Game ready! Click to launch ball');
    }
    
    init() {
        this.paddle = new FixedPaddle();
        this.ball = new FixedBall();
        this.ball.reset(this.paddle);
        this.bricks = [];
        this.particles = [];
        
        // Create bricks
        for (let row = 0; row < 6; row++) {
            for (let col = 0; col < 10; col++) {
                const x = 25 + col * 85;
                const y = 50 + row * 30;
                const type = Math.floor(row / 2) + 1;
                this.bricks.push(new FixedBrick(x, y, type));
            }
        }
        
        this.updateUI();
        console.log(`Created ${this.bricks.length} bricks`);
    }
    
    setupControls() {
        this.canvas.addEventListener('mousemove', (e) => {
            if (this.gameOver || this.paused) return;
            const rect = this.canvas.getBoundingClientRect();
            this.paddle.moveTo(e.clientX - rect.left);
        });
        
        this.canvas.addEventListener('click', () => {
            if (!this.gameOver && !this.paused) this.ball.launch();
        });
        
        document.addEventListener('keydown', (e) => {
            if (this.gameOver) return;
            
            switch(e.code) {
                case 'Space':
                    e.preventDefault();
                    if (!this.paused) this.ball.launch();
                    break;
                case 'KeyR':
                    this.restart();
                    break;
                case 'KeyP':
                    this.togglePause();
                    break;
                case 'Equal':
                case 'NumpadAdd':
                    if (!this.paused) this.ball.changeSpeed(1);
                    break;
                case 'Minus':
                case 'NumpadSubtract':
                    if (!this.paused) this.ball.changeSpeed(-1);
                    break;
            }
        });
    }
    
    start() {
        this.running = true;
        this.gameLoop();
    }
    
    update() {
        if (this.paused || this.gameOver) return;
        
        this.paddle.update();
        const ballStatus = this.ball.update(this.canvas, this.paddle, this.particles);
        
        // Ball-brick collisions
        for (let brick of this.bricks) {
            if (!brick.destroyed && this.ball.collides(brick)) {
                this.ball.dy = -this.ball.dy;
                const points = brick.hit(this.particles);
                this.score += points;
                this.updateUI();
                break;
            }
        }
        
        // Update particles
        this.particles = this.particles.filter(p => p.update());
        
        // Remove destroyed bricks
        this.bricks = this.bricks.filter(b => !b.destroyed);
        
        // Level complete
        if (this.bricks.length === 0) {
            this.nextLevel();
        }
        
        // Ball lost
        if (ballStatus === 'lost') {
            this.lives--;
            this.updateUI();
            
            if (this.lives <= 0) {
                this.endGame();
            } else {
                this.ball.reset(this.paddle);
                updateFixedStatus('Ball lost! Click to launch again');
            }
        }
    }
    
    draw() {
        // Clear with gradient background
        const gradient = this.ctx.createLinearGradient(0, 0, 0, this.canvas.height);
        gradient.addColorStop(0, '#001a2e');
        gradient.addColorStop(1, '#003d5c');
        this.ctx.fillStyle = gradient;
        this.ctx.fillRect(0, 0, this.canvas.width, this.canvas.height);
        
        // Draw stars
        for (let i = 0; i < 30; i++) {
            const x = (i * 137) % this.canvas.width;
            const y = (i * 211) % this.canvas.height;
            this.ctx.fillStyle = '#ffffff';
            this.ctx.fillRect(x, y, 1, 1);
        }
        
        // Draw game objects
        this.paddle.draw(this.ctx);
        this.ball.draw(this.ctx);
        
        this.bricks.forEach(brick => {
            if (!brick.destroyed) brick.draw(this.ctx);
        });
        
        this.particles.forEach(particle => particle.draw(this.ctx));
        
        // Pause overlay
        if (this.paused) {
            this.ctx.fillStyle = 'rgba(0, 0, 0, 0.7)';
            this.ctx.fillRect(0, 0, this.canvas.width, this.canvas.height);
            this.ctx.fillStyle = '#fff';
            this.ctx.font = '36px Arial';
            this.ctx.textAlign = 'center';
            this.ctx.fillText('PAUSED', this.canvas.width/2, this.canvas.height/2);
        }
    }
    
    nextLevel() {
        this.level++;
        this.ball.speed += 0.5;
        this.init();
        updateFixedStatus(`Level ${this.level}! Speed increased!`);
    }
    
    endGame() {
        this.gameOver = true;
        this.running = false;
        updateFixedStatus(`Game Over! Score: ${this.score}`);
    }
    
    restart() {
        this.score = 0; this.lives = 3; this.level = 1;
        this.gameOver = false; this.paused = false; this.running = false;
        this.init();
        this.start();
        updateFixedStatus('Game restarted!');
    }
    
    togglePause() {
        if (!this.gameOver) {
            this.paused = !this.paused;
            updateFixedStatus(this.paused ? 'Paused' : 'Resumed');
        }
    }
    
    updateUI() {
        document.getElementById('fixedScore').textContent = this.score;
        document.getElementById('fixedLives').textContent = this.lives;
        document.getElementById('fixedLevel').textContent = this.level;
        document.getElementById('fixedSpeed').textContent = this.ball.speed;
    }
    
    gameLoop() {
        if (this.running) {
            this.update();
            this.draw();
            requestAnimationFrame(() => this.gameLoop());
        }
    }
}

// Start fixed game
setTimeout(() => {
    try {
        window.fixedGame = new FixedGame();
        console.log('üéÆ Fixed Enhanced Game Ready!');
        console.log('üéØ Two-hit bricks + speed control working!');
    } catch (error) {
        updateFixedStatus('Error: ' + error.message);
        console.error('Game error:', error);
    }
}, 200);
</script>
```
